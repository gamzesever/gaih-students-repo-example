4) 
-Duplicated Values: Ver seti içindeki tekrar eden değerleri tespit etmeye yarar. Tekrar eden değerler modelin overfitting problemi yaşamasına neden olur. 
- Imbalanced Data: Bir sınıftaki gözlem sayısı diğer sınıf(lar)dan fazla olabilir. Bu durumda algoritma çalışırken, tüm gözlemleri fazla olan sınıfa atar ve bu da modelin doğruluğunu azaltır. Bu sorunu çözmek için sınıf sayıları eşitlenmedir. 
- Missing Values: Veri setindeki eksik olan gözlemlerdir. Bu sorunu çözmek için NA gözlemler veri setinden çıkarılabilir ya da ortalama/medyan değereri le doldurulabilir
- Outlier Detection: Veri setinin dağılımını etkiler ve yanlış model kurmamıza neden olabilir. 
- Feature Scaling: Değişkenlerin ortalama etrafında simetrik bir dağılım göstermesi ve Normal dağılım varsayımını sağlaması için Standartlaştırma ve Normalleştirme yöntemleri kullanılmaktadır. 
- Bucketing(binning): veri setindeki çok küçük gözlemlerden kaynaklanan hataları en aza indirmek için kullanılır. Bu yöntemde veri seti daha geniş aralıklara bölünerek modelin noise lardan etkilenmesi önlenir. 
- Feature Extraction: veri setindeki model için en anlamlı olan değişkenlerin belirlenmesidir. 
- Feature Encoding: verideki nominal ve ordinal verilerin {0,1,2,3} gibi numarik değerler ile ifade edilmesi için düzenlenmesidir. 
-Train/Validation/test split: verinin train-validation-test(0,2-0,2-0,6) ya da train-test(0,3-0-7) olarak ayrılmasıdır. 
Preprocessing, veri setindeki istenmeyen ve modeli olumsuz etkileyebilecek durumların ortadan kaldırılması için önemli bir adımdır. 

5)
-Continuous Variable: Herhangi bir aralıktaki tüm değerleri alabilen değişkenlerdir. Örneğin: boy uzunluğu. Bir değişkenin sürekli olup olmadığını anlamak için histogram grafiğine bakılabilir. 
-Discrete Variable: Bir aralık arasında sayılabilen numerik değerler alabilen verilerdir. Örneğin: yaş. Bir değişkenin kesikli olup olmadığını barchart grafiğine bakarak anlayabiliriz.
